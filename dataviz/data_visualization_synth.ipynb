```python
mod_df = mod_df[mod_df['vibration'] >= 0]
mod_df.info()
```

```python
mod_df_power = mod_df["motor_current"]
mod_df = mod_df.drop(columns=['motor_current'])
mod_df
```

```python
mod_df["failure"].value_counts()
```

```python
mod_df_failure = mod_df[mod_df["failure"] == 1].copy()
mod_df_success = mod_df[mod_df["failure"] == 0].copy()
mod_df_failure
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
```

```python
t0 = mod_df_failure['timestamp'].min()

mod_df_failure['t_sec'] = (mod_df_failure['timestamp'] - t0).dt.total_seconds()
features = ['t_sec', 'temperature', 'vibration', 'pressure', 'Power']
X = mod_df_failure[features].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
k_neighbors = 5
nn = NearestNeighbors(n_neighbors=k_neighbors + 1)
nn.fit(X_scaled)
neighbors_indices = nn.kneighbors(X_scaled, return_distance=False)
```

```python
original_count = X_scaled.shape[0]
target_failure_count = 4706 - original_count
n_to_generate = target_failure_count - original_count

synthetic_samples = []
np.random.seed(42)

for _ in range(n_to_generate):
    idx = np.random.randint(0, original_count)

    neighbor_idx = np.random.choice(neighbors_indices[idx][1:])
    lam = np.random.rand()
    synthetic_datapoint = X_scaled[idx] + lam * (X_scaled[neighbor_idx] - X_scaled[idx])
    synthetic_samples.append(synthetic_datapoint)

synthetic_samples = np.array(synthetic_samples)
```

```python
X_scaled_oversampled = np.vstack((X_scaled, synthetic_samples))
X_oversampled = scaler.inverse_transform(X_scaled_oversampled)

df_oversampled = pd.DataFrame(X_oversampled, columns=features)
df_oversampled['timestamp'] = t0 + pd.to_timedelta(df_oversampled['t_sec'], unit='s')
df_oversampled['failure'] = 1
df_oversampled.drop(columns=['t_sec'], inplace=True)
df_oversampled = df_oversampled[['timestamp', 'temperature', 'vibration', 'pressure', 'failure', 'Power']]
```

```python
print(original_count)
print(df_oversampled.shape[0])
```

```python
df_combined_failure = pd.concat([mod_df_failure, df_oversampled], axis=0)
df_combined_failure.drop(columns=['t_sec'], inplace=True)
df_combined_failure.shape[0]
```

```python
df_combined = pd.concat([df_combined_failure, mod_df_success], axis=0)
df_combined['failure'].value_counts()
```

```python
df_combined.to_csv("augmented_dataset.csv", index=False)
```
