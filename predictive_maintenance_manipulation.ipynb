```python
import pandas as pd

df = pd.read_csv('Predictive_Maintenance_Dataset.csv')
```

```python
power = []
for current in df['motor_current']:
    power.append(current * 220)

df['power'] = power
```

```python
df = df[['timestamp', 'temperature', 'vibration', 'pressure', 'motor_current', 'power', 'failure']]
```
